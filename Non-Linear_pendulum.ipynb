```python
# Considering the case when we have initial velocity = 0

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
n = float(input("Enter the angle in degree: "))
l = float(input("Enter the length of the wire: "))
T = float(input("Maximum time for the oscillation to continue: "))

theta_0 = ((np.pi)/180)*(n)   # in radian
g = 9.8 #m/s2
delta_t = 0.01
gbyl = g/l
# intial angular velocity = 0
# theta_1 = theta_0 - (g/l)(sin(theta_0))((\Delta t)^2)/2
```

```python
theta_1 = theta_0 - (gbyl*(np.sin(theta_0))*((delta_t)**2)/2)

time  = np.array([0       ,    0 + delta_t      ])
theta_array = np.array([theta_0 ,    theta_1    ])

i = 1
theta = theta_1
t = 0 + delta_t

while t < T:
    theta = 2*(theta_array[i]) - gbyl*(np.sin(theta_array[i]))*((delta_t)**2) - theta_array[i-1]   # this is theta_i+1
    theta_array = np.append(theta_array , theta)
    t = t + delta_t
    time = np.append(time , t)
    # t = t + delta_t
    i = i + 1
theta_final = 2*(theta_array[i]) - gbyl*(np.sin(theta_array[i]))*((delta_t)**2) - theta_array[i-1]

# theta in degree

angle_array = theta_array*(180/np.pi)
# i
```

```python
plt.plot(time , angle_array , label = "Non_linear SHM")
plt.legend()
plt.savefig("Non_linear SHM.png")
```
